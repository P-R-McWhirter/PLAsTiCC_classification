```python
import numpy as np
import pandas as pd
from scipy import interpolate
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
from gatspy import datasets, periodic
```

```python
# Read the data. Data are added with the "+ Add Data" button on the right hand panel.
# Look for the LSST competition
t = pd.read_csv("./training_set.csv")
m = pd.read_csv("./training_set_metadata_head.csv")
```

```python
# Get the array of Object ID and length
objectid_list = np.unique(m['object_id'])
objectid_list_len = len(objectid_list)
```

```python
# Loop 100 times
for j in range(1):

    # Draw 10 random Object IDs from the list
    objectid = objectid_list[np.int64(np.random.random() * objectid_list_len)]
    objectid = 76475797

    # Get the rows with the objectid drawn above
    lightcurve = t.loc[t['object_id']==objectid]

    # Data
    mjdstart = lightcurve['mjd'].iloc[0]
    u = lightcurve.loc[lightcurve['passband']==0]
    g = lightcurve.loc[lightcurve['passband']==1]
    r = lightcurve.loc[lightcurve['passband']==2]
    i = lightcurve.loc[lightcurve['passband']==3]
    z = lightcurve.loc[lightcurve['passband']==4]
    y = lightcurve.loc[lightcurve['passband']==5]

    # More data
    header = m.loc[m['object_id']==objectid]
    objclass = header['target'].iloc[0]
    specz = header['hostgal_specz'].iloc[0]
    photz = header['hostgal_photoz'].iloc[0]
    gallat = header['gal_b'].iloc[0]
    
    print(header)
    
    u_skew = scipy.stats.skew(u['flux'])
    
    g_skew = scipy.stats.skew(g['flux'])
    
    r_skew = scipy.stats.skew(r['flux'])
    
    i_skew = scipy.stats.skew(i['flux'])
    
    z_skew = scipy.stats.skew(z['flux'])
    
    y_skew = scipy.stats.skew(y['flux'])
    
    all_skew = [u_skew, g_skew, r_skew, i_skew, z_skew, y_skew]
    
    print(np.mean(all_skew))

    # Preset the title of the plot
    title = "Object number: " + str(objectid) + ", Object class: " + str(objclass) + "\n\n PhotZ = " + str(photz) + ", SpecZ = " + str(specz) + ", GalLat = " + str(gallat) + "deg"

    # Configure plot
    f = plt.figure(figsize=(12,12))
    plt.subplots_adjust(hspace=0)

    # Adding a subplot to the 6th(third argument) plotting environment of size 6 (first) down 1 (second) across
    ax1 = f.add_subplot(6,1,6)
    plt.ylabel("Y")
    plt.errorbar(y['mjd']-mjdstart, y['flux'], yerr=y['flux_err'], fmt='o',color='brown')
    plt.grid()

    ax2 = f.add_subplot(6,1,5,sharex=ax1)
    plt.ylabel("z")
    plt.errorbar(z['mjd']-mjdstart, z['flux'], yerr=z['flux_err'], fmt='o',color='red')
    plt.grid()

    ax3 = f.add_subplot(6,1,4,sharex=ax1)
    plt.ylabel("i")
    plt.errorbar(i['mjd']-mjdstart, i['flux'], yerr=i['flux_err'], fmt='o',color='orange')
    plt.grid()

    ax4 = f.add_subplot(6,1,3,sharex=ax1)
    plt.ylabel("r")
    plt.errorbar(r['mjd']-mjdstart, r['flux'], yerr=r['flux_err'], fmt='o',color='green')
    plt.grid()

    ax5 = f.add_subplot(6,1,2,sharex=ax1)
    plt.ylabel("g")
    plt.errorbar(g['mjd']-mjdstart, g['flux'], yerr=g['flux_err'], fmt='o',color='blue')
    plt.grid()

    ax6 = f.add_subplot(6,1,1,sharex=ax1)
    plt.ylabel("u")
    plt.errorbar(u['mjd']-mjdstart, u['flux'], yerr=u['flux_err'], fmt='o',color='purple')
    plt.grid()

    plt.title(title)
    plt.xlabel("MJD")
```

```python
model = periodic.LombScargleMultibandFast(fit_period=True)

baseline = np.max(lightcurve['mjd']) - np.min(lightcurve['mjd'])

model.optimizer.period_range=(0.05, baseline)

model.fit(lightcurve['mjd'], lightcurve['flux'], lightcurve['flux_err'], lightcurve['passband'])

model.best_period

power = model.score(model.best_period)
```

```python
dy_u = np.mean(u['flux_err'])
dy_g = np.mean(g['flux_err'])
dy_r = np.mean(r['flux_err'])
dy_i = np.mean(i['flux_err'])
dy_z = np.mean(z['flux_err'])
dy_y = np.mean(y['flux_err'])

print(model.best_period)

print(power)

gp_u = interpolate.interp1d(np.append(np.append(0, np.asarray(u['mjd']-mjdstart)), 1200), np.append(np.append(0, np.asarray(u['flux'])), 0), fill_value="extrapolate")
gp_g = interpolate.interp1d(np.append(np.append(0, np.asarray(g['mjd']-mjdstart)), 1200), np.append(np.append(0, np.asarray(g['flux'])), 0), fill_value="extrapolate")
gp_r = interpolate.interp1d(np.append(np.append(0, np.asarray(r['mjd']-mjdstart)), 1200), np.append(np.append(0, np.asarray(r['flux'])), 0), fill_value="extrapolate")
gp_i = interpolate.interp1d(np.append(np.append(0, np.asarray(i['mjd']-mjdstart)), 1200), np.append(np.append(0, np.asarray(i['flux'])), 0), fill_value="extrapolate")
gp_z = interpolate.interp1d(np.append(np.append(0, np.asarray(z['mjd']-mjdstart)), 1200), np.append(np.append(0, np.asarray(z['flux'])), 0), fill_value="extrapolate")
gp_y = interpolate.interp1d(np.append(np.append(0, np.asarray(y['mjd']-mjdstart)), 1200), np.append(np.append(0, np.asarray(y['flux'])), 0), fill_value="extrapolate")

#maxtime_u = np.max(u['mjd']-mjdstart)
#maxtime_g = np.max(g['mjd']-mjdstart)
#maxtime_r = np.max(r['mjd']-mjdstart)
#maxtime_i = np.max(i['mjd']-mjdstart)
#maxtime_z = np.max(z['mjd']-mjdstart)
#maxtime_y = np.max(y['mjd']-mjdstart)

#maxtime = np.max([maxtime_u, maxtime_g, maxtime_r, maxtime_i, maxtime_z, maxtime_y])

x = np.linspace(0, 1200, 2000)
```
